## Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
data = load_breast_cancer()
y = data["target"]
data = pd.DataFrame(data = data["data"], columns=data["feature_names"])
```

```python
_x = np.linspace(0, 10, 100)
_y = _x**2
plt.plot(_x, _y)
```

```python
plt.scatter(data.loc[y==1, "mean radius"], data.loc[y==1, "worst perimeter"]) # mexer com opções
```

```python
for i in {0, 1}:
    plt.scatter(data.loc[y==i, "mean radius"], data.loc[y==i, "worst perimeter"])
```

```python
for i in {0, 1}:
    plt.scatter(data.loc[y==i, "mean radius"], [0 for _ in range(np.count_nonzero(y==i))], alpha = .5 )
```

```python
plt.hist(data.loc[:, "mean radius"], bins = 50);
```

```python
for i in {0, 1}:
    plt.hist(data.loc[y == i, "mean radius"], bins = 50, alpha = .5);
```

```python
plt.figure(figsize=(5, 5))
plt.scatter([0, 1], [0, 1], s = 100, marker = "X")
plt.scatter([0, 1], [1, 0], s = 100)
```
